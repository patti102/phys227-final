### Display of all plotting functions using c = 2

Here we demonstrate all of the plotting functions by using a non-chaotic $c$ value parameter of $c = 2$. These come from solving the differential equations:

$$\dot{x} = -y -z $$
$$\dot{y} = x = ay$$
$$\dot{z} = b + z(x - c)$$

and then integrating these functions via the 4th-Order Runga-Kutta method. First, we plot each of the three function values against time, which illustrates how they rapidly reach steady state and then fall into repetitive, harmonic motion.

```python
value2 = p1.Rossler(2)
value2.run()
value2.plotx()
```

```python
value2.ploty()
```

```python
value2.plotz()
```

We now show the 2 and 3-D plots, which illustrate combinations of these functions. Unlike the 1-D plots, these plots only show data from the second half of the integration period. This is to exclude the data points which are taken before the system reaches a steady state. This way, we only see the oscilatory behavior, fixed point behavior, or at the very least, chaotic behavior which is clearly chaotic behavior and not just the initial non-harmonic initial noise. Here we have oscillatory behavior. That is why we only get a single path around for each plot. The system of equations is completely periodic with only one type of cycle at this value of $c$.

### Bifurcation Diagrams of Local Maximums

By isolating and plotting all of the local maximum values vs the c component used to obtain them, we create one of those famous bifurcation diagrams. In the midterm, we simply used the logistic equation, which returns only one point and thus can be plotted directly. Here, we must establish criteria for the type of point which we wish to obtain from each cycle. Selecting the local maxima or minima is a good and simple choice as it is easily quantified and it occurs precisely once per cycle.

The three plots below show the bifurcation diagrams of x, y, and z respectively. They all display the hallmark features of such a plot, such as period doubling, an eventual period cascade, and even regions of non-chaotic behavior at parameter values higher than regions which do have chaotic behavior. While none of these features are new or surprising to us, it is somewhat instructive to consider the differences between the bifurcation diagrams of the three different functions. Notice that, while all three diagrams display the same general behavior and bifurcate at the same c values, the precise values and ranges of their max values differs. The bifurcation at the same c values can only be expected, considering that the bifurcation of one equation would cause the bifurcation of another as they are all coupled in their differential equations. The differences in max values is likewise expected, as we have already witnessed that the functions used to obtain these diagrams differ from one another in their 1 and 2-dimensional time parametric plots.
